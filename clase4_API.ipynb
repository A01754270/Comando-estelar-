# Implementacion de API

```python
import pandas as pd

# Datos prueba
data_test =  pd.read_csv("datos_prueba.csv")
```

```python
# Se elimina la columna correspondiente a la fecha de modificación (target) si existe
target = "ModifiedDate"
if target in data_test.columns:
    data_test = data_test.drop(columns=[target])
```

```python
# Se convierte cada fila del dataset en un diccionario
data_dict = data_test.to_dict(orient='records') 
```

```python
import json

# El dataset se formatea a una lista de diccionarios para utilizarse en la API
data_json = json.dumps({"data": data_dict})
# print(data_json)  
```

```python
# Obtención del scoring uri
suri = open("uri.json", "r")
scoring_uri = json.load(suri)["URI"][0]
suri.close()
```

```python
import requests

headers = {"Content-Type": "application/json"}

# Request a la API
response = requests.post(scoring_uri, data=data_json, headers=headers)

# Comprobando el status de la respuesta y obteniendo el resultado
if response.status_code == 200:
  result = json.loads(response.json())
  print("Predicción:", result)
  data_test["ModifiedDate Predicted"] = result   # Añadiendo la predicción al dataset
  display(data_test)
else:
  print(f"Error: {response.text}")
```
